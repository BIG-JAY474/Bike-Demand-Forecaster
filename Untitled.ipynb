```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
# Set a nice visual style
sns.set_style('whitegrid') 
```

```python
df = pd.read_csv("day.csv")
print(f"Original Data Shape: {df.shape}")
print(f"Number of missing values: {df.isnull().sum().sum()}")
```

```python
# 2. Rename columns for easier understanding (optional but recommended)
df.rename(columns={
    'dteday': 'date', 
    'yr': 'year', 
    'mnth': 'month', 
    'weathersit': 'weather_condition', 
    'hum': 'humidity', 
    'cnt': 'total_rentals'
}, inplace=True)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
categorical_cols = ['season', 'year', 'month', 'holiday', 'weekday', 'workingday', 'weather_condition']
for col in categorical_cols:
    df[col] = df[col].astype('category')

print("\nData Types after preparation:")
print(df.dtypes)
```

```python
df.info()
```

```python
# Visualization 1: Time Series Plot of Total Rentals (Before and After the year change)
plt.figure(figsize=(15, 6))
sns.lineplot(data=df, x='date', y='total_rentals')
plt.title('Total Bike Rentals Over Time (Raw Data Visualization)')
plt.xlabel('Date')
plt.ylabel('Total Rentals')
plt.show()
```

```python
# Visualization 2: Correlation Heatmap
# Select numerical columns and our target variable
numerical_cols = ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'total_rentals']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features and Total Rentals')
plt.show()
```

```python
# Visualization 3: Average Rentals by Weather Condition (Detailed EDA)
plt.figure(figsize=(8, 5))
avg_rentals_by_weather = df.groupby('weather_condition')['total_rentals'].mean().sort_values()
sns.barplot(x=avg_rentals_by_weather.index, y=avg_rentals_by_weather.values, palette='viridis')
plt.xticks(ticks=[0, 1, 2], labels=['Clear/Cloudy (1)', 'Misty/Rain (2)', 'Light Snow/Rain (3)'])
plt.title('Average Daily Rentals by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Average Total Rentals')
plt.show()
```

```python
# Visualization 4: Average Rentals by Month (Detailed EDA)
plt.figure(figsize=(10, 6))
avg_rentals_by_month = df.groupby('month')['total_rentals'].mean()
sns.barplot(x=avg_rentals_by_month.index, y=avg_rentals_by_month.values, palette='RdYlBu')
plt.title('Average Daily Rentals Per Month')
plt.xlabel('Month')
plt.ylabel('Average Total Rentals')
plt.show()
```

```python
# Visualization 5: Relationship between Temperature and Rentals (Detailed EDA)
df['temp_denorm'] = df['temp'] * 41 # Denormalize temp for better plot labels
plt.figure(figsize=(8, 6))
sns.regplot(x='temp_denorm', y='total_rentals', data=df, scatter_kws={'alpha':0.5, 's':20}, line_kws={'color':'red'})
plt.title('Total Rentals vs. Actual Temperature (C)')
plt.xlabel('Temperature (C)')
plt.ylabel('Total Rentals')
plt.show()
```

```python
# Extracting month and day of week as useful features for the model
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek
df['day_name'] = df['date'].dt.day_name() # e.g., 'Monday', 'Tuesday'
```

```python
# Create an 'is_weekend' flag as a binary feature
df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)

print("\nNew Features Added:")
print(df[['date', 'month', 'day_of_week', 'day_name', 'is_weekend']].head())

# --- NEW FEATURE: is_holiday_week ---

# Define the dates of Christmas Day and New Year's Day (which are already marked in 'holiday' column)
# We need to find the range of dates around them where demand drops off

# Function to check if a date falls within a 'holiday week' window (Dec 22nd to Jan 2nd)
def is_holiday_season(date):
    if (date.month == 12 and date.day >= 22) or (date.month == 1 and date.day <= 2):
        # We also need to make sure it wasn't a standard 'working day' that week (e.g. if we had 3 years of data)
        # For our 2-year dataset, this simple date check is sufficient to capture the drop in commute behavior
        return 1 
    return 0
    # Apply the function to our date column to create the new feature
df['is_holiday_week'] = df['date'].apply(is_holiday_season)

# Convert the new feature to a category type (optional but good practice for XGBoost)
df['is_holiday_week'] = df['is_holiday_week'].astype('category')

print("\nAdded the new 'is_holiday_week' feature:")
print(df[['date', 'is_holiday_week', 'total_rentals']].tail(12))
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# 1. Define Features (X) and Target (y)
# We exclude 'date', 'casual', 'registered', and redundant feature names
features = ['season', 'year', 'month', 'holiday', 'weekday', 'workingday', 
            'weather_condition', 'temp', 'humidity', 'windspeed', 'is_weekend', 'is_holiday_week']
target = 'total_rentals'

X = df[features]
y = df[target]
```

```python
# 2. Perform Time-Series Split (We split the last 60 days for testing/validation)
# In a robust system, we would use time-series cross validation, 
# but for a basic project, a simple cutoff works:
cutoff_date = pd.to_datetime('2012-10-31') # Roughly 60 days before the end of the data

X_train = X[df['date'] <= cutoff_date]
y_train = y[df['date'] <= cutoff_date]
X_test = X[df['date'] > cutoff_date]
y_test = y[df['date'] > cutoff_date]

print(f"\nTraining set size: {len(X_train)} days")
print(f"Testing set size: {len(X_test)} days")
```

```python
model = XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.05, random_state=42, enable_categorical=True)
```

```python
print("\nStarting model training...")
model.fit(X_train, y_train)
print("Model training complete.")
```

```python
from xgboost import plot_importance
plt.figure(figsize=(10, 8))
plot_importance(model, importance_type='weight', title='Feature Importance for Demand Forecaster')
plt.xlabel('Frequency/Weight of Feature Use in Model Decisions')
plt.show()
```

```python
# 1. Generate Predictions on the Test Set
predictions = model.predict(X_test)
results_df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
results_df['date'] = df[df['date'] > pd.to_datetime('2012-10-31')]['date'].values
print("\nFirst 10 Actual vs Predicted Values:")
print(results_df.head(10))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)
print(f"\n--- Model Performance Metrics (Test Set) ---")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared Score (R^2): {r2:.4f}")
```

```python
# Visualization 6: Actual vs. Predicted Time Series Plot (Core Evaluation Plot)
plt.figure(figsize=(15, 6))
sns.lineplot(data=results_df, x='date', y='Actual', label='Actual Rentals', color='blue')
sns.lineplot(data=results_df, x='date', y='Predicted', label='Predicted Rentals', color='orange', linestyle='--')
plt.title('Demand Forecaster: Actual vs. Predicted Rentals on Test Data')
plt.xlabel('Date')
plt.ylabel('Total Rentals')
plt.legend()
plt.show()
```

```python
# Visualization 7: Residual Plot (Error Analysis)
residuals = y_test - predictions
results_df['Residuals'] = residuals
plt.figure(figsize=(15, 6))
sns.scatterplot(x='date', y='Residuals', data=results_df, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', label='Zero Error Line')
plt.title('Residual Analysis: Errors Over Time')
plt.xlabel('Date')
plt.ylabel('Prediction Error (Residuals)')
plt.legend()
plt.show()
```
